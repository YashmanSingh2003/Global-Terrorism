# Comparing Terror attacks with respect to years

**As we can see the rete of terror attacks kept increasing and it was at the peak during 2012-2015** 

```python
x=list(new_df['Year'])
from collections import Counter
datam = Counter(x)
print(datam.most_common()[:3])
sns.displot(x,height=6)
plt.xlabel('Years')
plt.ylabel('No. of attack')
plt.title('Rate of teror attacks')
plt.show()
```

# Terror attacks at different countries

```python
coun_terror=new_df['Country'].value_counts()[:15].to_frame()
coun_terror.columns=['Attracks']
coun_kill=new_df.groupby('Country')['people died'].sum().to_frame()
coun_terror.merge(coun_kill,left_index=True,right_index=True,how='left').plot.bar(width=0.9)
fig=plt.gcf()
fig.set_size_inches(16,6)
plt.title('Terror prone countries')
plt.show()
```

# Terror attack at different cities

# Most common form of attacks

```python
x=list(new_df['Attact type'])
from collections import Counter
datam = Counter(x)
print(datam.most_common()[:3])
sns.displot(x,height=19)
plt.xlabel('Methord')
plt.ylabel('No. of attack')
plt.title('Most common form of attack')
plt.xticks(rotation = 45)
plt.show()
```

# Most common attack targets

```python
x=list(new_df['targtype1_txt'])
from collections import Counter
datam = Counter(x)
print(datam.most_common()[:3])
sns.displot(x,height=19)
plt.xlabel('Target')
plt.ylabel('No. of attack')
plt.title('Most common target of attacks')
plt.xticks(rotation = 90)
plt.show()
```
